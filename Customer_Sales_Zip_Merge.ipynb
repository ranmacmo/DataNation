```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
```

```python
# Load the customer dataset
file_path = "data/Customer_sales.csv"
df_customer_sales = pd.read_csv(file_path,dtype={"BillPostalCode":"string"} )
df_customer_sales
```

```python
# Load zipcode dataset and ensure that zipcodes are read as strings
file_path = "data/USZip_info.csv"
df_zipcodes= pd.read_csv(file_path,dtype={"zip":"string"})
df_zipcodes.head(10)
```

```python

df_customer_sale_demographics  = pd.merge(df_customer_sales, df_zipcodes, left_on='BillPostalCode', right_on='zip')
df_customer_sale_demographics
```
